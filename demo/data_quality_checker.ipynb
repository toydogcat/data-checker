# 📊 資料集有效性驗證工具 (Data Quality Checker)

本 Notebook 提供全面的資料品質與健康度體檢流程，包含三個階段：
1. **階段一**：自動化統計、結構健康度（缺失率、重複率、格式）、內容資訊豐富度（字數、Entropy、TTR）與安全合規（PII 個資與毒性過濾）。
2. **階段二**：使用 Embedding（向量化）評估多樣性與話題涵蓋廣度。
3. **階段三**：引入「Llama / LLM as a Judge」抽樣審查（流暢度、邏輯性、資訊價值）。

## 🛠️ 1. 環境設定與套件載入 (conda environment: `toby` / Colab)

```python
# 在 Colab 環境中若需要安裝 optional 套件可取消註解執行：
# !pip install pandas numpy scikit-learn matplotlib seaborn sentence-transformers openai

import os
import sys
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 確保引進 data_checker 模組
sys.path.append('..')
from data_checker import Phase1Checker, Phase2Checker, Phase3Checker

print('✅ Data Quality Checker Toolkit 載入成功！')
```

## 📈 2. 階段一：自動化統計與結構健康度體檢 (Phase 1)

```python
# 2.1 載入測試用的 DataFrame 數據
df_path = './dataframe_dataset.csv'
df = pd.read_csv(df_path)
print(f"載入 DataFrame 資料筆數: {len(df)}, 欄位數: {len(df.columns)}")

# 2.2 執行 Phase 1 體檢
p1_checker = Phase1Checker()
df_report = p1_checker.analyze_dataframe(df)

print(f"=== 階段一數據體檢報告 (品質得分: {df_report['quality_score']} / 100) ===")
print(f"- 整體缺失率: {df_report['overall_missing_rate']}%")
print(f"- 資料重複率: {df_report['duplication_rate']}%")
print(f"- PII 敏感個資總筆數: {df_report['text_analysis'].get('total_pii_found', 0)}")
print(f"- 毒性與不當字詞筆數: {df_report['text_analysis'].get('toxic_records_count', 0)}")
print(f"- 詞彙多樣性 (TTR): {df_report['text_analysis'].get('ttr', 0)}")
```

```python
# 2.3 繪製各欄位缺失率條狀圖
col_missing = df_report['column_missing_rates']
col_names = list(col_missing.keys())
missing_rates = [col_missing[c]['missing_rate'] for c in col_names]

plt.figure(figsize=(10, 4))
plt.barh(col_names, missing_rates, color='#e74c3c')
plt.xlabel('Missing Rate (%)')
plt.title('Column Sparsity / Missing Rates')
plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## 🌐 3. 階段二：Embedding 多樣性與話題廣度評估 (Phase 2)

```python
# 3.1 載入文本 JSONL 語料庫數據
text_path = './text_dataset.jsonl'
texts = []
with open(text_path, 'r', encoding='utf-8') as f:
    for line in f:
        if line.strip():
            item = json.loads(line)
            texts.append(item['text'])

print(f"載入語料庫總筆數: {len(texts)}")

# 3.2 執行 Phase 2 多樣性評估
p2_checker = Phase2Checker()
p2_report = p2_checker.evaluate_diversity(texts)

print(f"=== 階段二 Embedding 多樣性報告 ===")
print(f"- 向量計算方法: {p2_report['embedding_method']}")
print(f"- 平均兩兩餘弦距離 (Distance): {p2_report['mean_pairwise_distance']}")
print(f"- 平均餘弦相似度 (Similarity): {p2_report['avg_cosine_similarity']}")
print(f"- 前三維 PCA 主成分解釋變異數: {p2_report['top3_pca_explained_variance']}")
print(f"- 多樣性綜合得分: {p2_report['diversity_score']} / 100")
print(f"- 評定結論: {p2_report['interpretation']}")
```

## ⚖️ 4. 階段三：Llama as a Judge 抽樣審查 (Phase 3)

```python
# 4.1 執行 Phase 3 LLM Judge 抽樣評分 (無 API key 時自動啟用 Mock/Heuristic 評估引擎)
p3_checker = Phase3Checker()  # 若有 OpenAI API KEY 可傳入 api_key="..."
p3_report = p3_checker.audit_corpus(texts, sample_size=10)

print(f"=== 階段三 LLM Judge 評分報告 (執行模式: {p3_report['execution_mode']}) ===")
print(f"- 平均流暢度 (Fluency): {p3_report['average_scores']['fluency']} / 5.0")
print(f"- 平均邏輯性 (Logic):   {p3_report['average_scores']['logic']} / 5.0")
print(f"- 平均資訊價值 (Value): {p3_report['average_scores']['value']} / 5.0")
print(f"- 綜合平均得分:          {p3_report['average_scores']['overall_score']} / 5.0")
print(f"- 通過門檻評定 (>=4.0):  {'✅ 通過' if p3_report['judge_pass'] else '❌ 未通過'}")

# 展示前 3 筆抽樣結果
sample_df = pd.DataFrame(p3_report['sample_evaluations'])
sample_df[['sample_index', 'text_snippet', 'fluency', 'logic', 'value', 'reason']].head(5)
```
